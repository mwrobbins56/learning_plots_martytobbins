```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
# Group by genre and calculate the average total streams
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.figure(figsize=(10,6))
colors = ["#1c9099"]*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#dd1c77"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

special_artist = "Ariana Grande"
colors = ["#31a354" if artist == special_artist else "purple" for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.9)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Ariane Grande 2nd in Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"],
    color = "blue",
    alpha = 0.9,
    s = 25)

plt.xlabel("Streams This Month")
plt.ylabel("Total Streams")
plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

featured_genre = "Pop"

colors = ["green" if genre == featured_genre else "pink" for genre in spotify_df["Genre"]]

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"],
    color = colors,
    alpha = 0.9,
    s = 25)

# plt.xlabel("Streams This Month")
plt.ylabel("Total Streams")
plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve Higher Monthly Listens or Total Streams?", fontsize = 18, pad = 45)

for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```

```python
# Line Chart
yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(12,8))
plt.plot(
    yearly_avg.index, 
    yearly_avg.values,
    marker = "+",
    linestyle = '--',
    color = "#dd1c77",
    linewidth = 1)

plt.xlabel("Release Year")
plt.ylabel("Average Total Streams")
plt.title("Did Listeners Stream More in 2023?", fontsize = 18, pad = 30)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.axvline(
    2022, 
    color = "#8856a7",
    linewidth = 0.9,
    alpha = 0.7)

ax.text(
    2022+0.05,
    2025,
    "Dip in Total Streams",
    color = "#2b8cbe",
    fontsize = 10
)

plt.ylim(0, 3500)

plt.show()
```

```python
# Pie Chart
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    platform_totals,
    labels=None,
    startangle=90,
    colors = ["#e34a33", "#addd8e"],
    autopct = "%1.1f%%")

plt.title("Split between Prmium and Free Users")

plt.legend(
    labels = platform_totals.index,
    loc = "lower right",
    title = "Platforms"
)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

wedges, texts, autotexts = ax.pie(
    platform_totals,
    autopct = "%1.1f%%",
    startangle=90,
    colors = ["#e34a33", "#addd8e"],)

ax.set_title(
    "Split Between Free and Premium Users",
    fontsize = 18,
    pad = 20,
    weight = "bold",
    loc = "center"
)

for autotext in autotexts:
    autotext.set_fontsize(30)
    autotext.set_color("yellow")
    autotext.set_weight("bold")

plt.legend(
    wedges,    
    platform_totals.index,
    loc = "lower center",
    bbox_to_anchor = (0.5, 0.95),
    ncol = 2
)

plt.show()
```

```python
# Histograms
plt.figure(figsize=(12,8))
plt.hist(
    spotify_df["Avg Stream Duration (Min)"],
    bins = 20,
    color = "#e34a33",
    alpha = 0.8,
    edgecolor = "white"
)

plt.xlabel("Average Stream Duration (Minutes)")
plt.ylabel("Number of Tracks")
plt.title("Distributation of Average Stream Duration", fontsize = 15, pad = 25)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.show()
```

```python
# Boxplots
plt.boxplot(
    [spotify_df[spotify_df["Platform Type"]=="Free"]["Skip Rate (%)"], 
    spotify_df[spotify_df["Platform Type"]=="Premium"]["Skip Rate (%)"]]
)

plt.xticks([1,2],["Free", "Premium"])
plt.xlabel("Plateform Type")
plt.ylabel("Skip Rate (%)")
plt.title("Skip Rate (%) by Platform Type")

plt.show()
```

```python
# Heat Maps
numeric_spotify_df = spotify_df.select_dtypes(include=[np.number])
corr = numeric_spotify_df.corr()

im = plt.imshow(corr, vmin = -1, vmax = 1, cmap = "coolwarm")
plt.colorbar(im, label="Pearson Corellation")

plt.xticks(range(len(corr.columns)), corr.columns, rotation = 45, ha = "right")
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Heat Map")

plt.show()
```

```python
# Count plots
genre_counts = spotify_df["Genre"].value_counts().sort_values(ascending=False)
genre_counts

plt.figure(figsize=(12,8))
plt.bar(
    genre_counts.index, 
    genre_counts.values,
    color = "#e34a33",
    alpha = 0.8,
    edgecolor = "white"
)

plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Number of Tracks by Genre - Countplots")
plt.xticks(rotation=45, ha="right")

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.show()
```
